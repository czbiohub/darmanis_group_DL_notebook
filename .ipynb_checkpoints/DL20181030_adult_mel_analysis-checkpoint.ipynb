# Global expression statistics
- Analysis read count and gene coverage 

```python
# Determine degree of expression level explained by genes (count>0) per cell
    # sum of counts as proxy for overall cell expression level
    
n_genes = pre_adata.iloc[:,1:].apply(lambda x: x>0).sum()
med_count = pre_adata.iloc[:,1:].replace(0, np.nan).median() # non-zero median
n_count = pre_adata.iloc[:,1:].sum()

df = pd.DataFrame({'n_genes':n_genes,'n_count':n_count, 'med_count':med_count})
print(df.corr())

plotnine.options.figure_size = (5,5)
plot_df = pd.melt(df, id_vars='n_genes')
(ggplot(plot_df, aes('value','n_genes'))+
    theme_bw()+
    theme(aspect_ratio=1)+
    geom_point()+
    scale_x_log10()+
    scale_y_log10()+
    facet_wrap('~variable', nrow=1, scales='free')+
    labs(y='# of genes observed'))
```

# Cohort analysis
- Determine general statistics
- Determine feature class statistics

```python
input_data = raw_adata

# Cell distribution among patients
plot_data = list(input_data.obs.groupby('patient').count()['count_idx'])
df = pd.DataFrame({'counts':plot_data})

plotnine.options.figure_size = (2,2)
print(ggplot(df, aes(x='counts'))+
     theme_bw() +
     theme(aspect_ratio=1)+
    geom_histogram()+
    labs(x='cell count', y='frequency'))

# Patient distribution among classes
count_list = []
label_list = []
group_list = []
for feat in ['age','color','sex','race','location']:
    count_list = count_list + input_data.obs.groupby(by=feat).nunique()['patient'].values.tolist()
    labels = input_data.obs.groupby(by=feat).nunique().index.tolist()
    label_list = label_list + labels
    group_list = group_list + [feat for x in range(len(labels))]
df = pd.DataFrame({'group':group_list,'key':label_list,'count':count_list})

plotnine.options.figure_size = (10, 10)
print(ggplot(df, aes('key','count'))+
     theme_bw() +
     theme(aspect_ratio=1,
          axis_text_x=element_text(angle=90),
          panel_spacing_x=0.3) +
    geom_bar(stat='identity', position='dodge') +
    facet_wrap('~group', scales='free', nrow = 1) +
    labs(y='pateint count', x=''))
```

# Cluster occupancy by patient 

```python
patients = adata_subset2.obs['patient'].tolist()
clusters = adata_subset2.obs['louvain'].tolist()

# determine counts per cluster by patient class
df = pd.DataFrame({'cluster':clusters, 'patients':patients})
df = df.groupby(['cluster','patients']).size()
df = df.unstack().replace(np.nan, 0)
df = df.reset_index()
print(df.to_string(index=False))
df = pd.melt(df, id_vars='cluster')

# determine class balance using CV as metric
stats_df = pd.merge(df.groupby('cluster').mean().reset_index(), df.groupby('cluster').std().reset_index(), on='cluster')
stats_df['CV'] = stats_df.value_y/stats_df.value_x
stats_df.columns = ['cluster','mean','std','CV']
print(stats_df.to_string(index=False))

plotnine.options.figure_size = (4,4)
(ggplot(df, aes('cluster','value',fill='patients'))+
     theme_bw()+
     theme(aspect_ratio=1)+
     geom_bar(stat='identity')+
     labs(y='# of cells'))
```

# Patient cell distribution by cluster

```python
# determine counts per patient by cluster class
df = pd.DataFrame({'cluster':clusters, 'patients':patients})
df = df.groupby(['patients','cluster']).size()
df = df.unstack().replace(np.nan, 0)
df = df.reset_index()
df.iloc[:,1:] = df.iloc[:,1:].apply(lambda x: x/sum(x), axis=1) # fraction of sum
print(df.to_string(index=False))
df = pd.melt(df, id_vars='patients')

plotnine.options.figure_size = (4,4)
(ggplot(df, aes('patients','value',fill='cluster'))+
     theme_bw()+
     theme(aspect_ratio=1,
          axis_text_x=element_text(angle=90))+
     geom_bar(stat='identity')+
     labs(y='Fractional cell frequency'))
```

# Quantify class covariance
- High degree of covariation indicates dependent classes

```python
# correlation among features
input_data = adata_subset2

fltrd_labels = [key for key,value in anno_dict.items()] + ['louvain']
df = pd.DataFrame()   
for label_col in fltrd_labels:
    acc_list = []
    for label_row in fltrd_labels:
        acc_list.append(class2class_reg(X=input_data.obs[label_row], y=input_data.obs[label_col]))
    df['{}'.format(label_col)] = acc_list
df['predictor'] = fltrd_labels
# print(df)

plotnine.options.figure_size = (3,3)
plot_df = pd.melt(df, id_vars=['predictor'])
(ggplot(plot_df, aes('variable','predictor',fill='value')) +
     theme_bw() +
     theme(aspect_ratio=1,
          axis_text_x=element_text(angle=90)) +
     geom_tile() +
     labs(x='response', y='predictor', fill='F1-score')+
    scale_fill_cmap(name='afmhot'))
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances

### create pc vs gene list heat map to determine correlated loading accross PCs
input_adata = adata_subset
pc_array = np.zeros((len(df),input_adata.varm[0][0].shape[0]))
gene_list = df['0'].tolist()

for row, target in enumerate(gene_list):
    pc_array[row,:] = input_adata[:,target].varm['PCs'].copy()[0]

# truncate and scale data
pc_array = pc_array[:,:10]
pc_array = preprocessing.scale(pc_array)
    
# cluster array
for n in range(2,10,1):
#     clustering = AgglomerativeClustering(n_clusters=n).fit(pc_array)
    clustering = KMeans(n_clusters=n, random_state=0).fit(pc_array)
    labels = clustering.labels_
    print(metrics.silhouette_score(pc_array, labels, metric='euclidean'))
    cluster_df = pd.DataFrame({'gene':gene_list,
                              'label':labels})

    # recast data
    plot_df = pd.DataFrame(data = pc_array, columns=['PC_{}'.format(x) for x in range(pc_array.shape[1])])
    plot_df['gene'] = gene_list
    plot_df = pd.melt(plot_df, id_vars='gene')

    # create a categorical type
    pc_cat = CategoricalDtype(categories=['PC_{}'.format(x) for x in range(pc_array.shape[1])], ordered=True)
    gene_cat = CategoricalDtype(categories=cluster_df.sort_values('label')['gene'].tolist(), ordered=True)

    # Cast the existing categories into the new category.
    plot_df['pc_cat'] = plot_df['variable'].astype(str).astype(pc_cat)
    plot_df['gene_cat'] = plot_df['gene'].astype(str).astype(gene_cat)

    plotnine.options.figure_size = (10,10)
    print(ggplot(plot_df, aes(x='pc_cat', y='gene_cat', fill='value'))+
        theme_bw()+
         theme(aspect_ratio=1,
              axis_text_x=element_text(angle=90,size=6),
              axis_text_y=element_text(size=6))+
         geom_tile())
```

```python

### DEV ###

def de_to_csv(input_adata):
    # Create DE output file
    # Input: ad obj that contains 'uns' array = 'rank_genes_groups'
    # Output: csv of DE

    x = input_adata.uns['rank_genes_groups']['logfoldchanges']
    fold_array = x.view('<f4').reshape(len(x),-1)
    print(fold_array[:5,:])

    x = input_adata.uns['rank_genes_groups']['names']
    names_array = x.view('<U50').reshape(len(x),-1)
    print(names_array[:5,:])
    
    num_cols = names_array.shape[1]
    
    df1 = pd.DataFrame({'name':names_array[:,0].tolist(),
                        'c_0': fold_array[:,0].tolist()})
    
    for col in range(1,num_cols,1):
        new_names = names_array[:,col].tolist()
        new_folds = fold_array[:,col].tolist()
        new_col = 'c_{}'.format(col)
        
        df2 = pd.DataFrame({'name':new_names, new_col:new_folds})
        print(df2.head())
        df1 = pd.merge(df1, df2, on='name')
        
    return df1

df = de_to_csv(adata_subset)
```
