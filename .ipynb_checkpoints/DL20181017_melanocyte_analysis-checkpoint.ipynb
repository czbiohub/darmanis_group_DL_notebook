```python
# base
import pandas as pd
import numpy as np
from scipy import sparse
import warnings


# sc analysis
import scanpy.api as sc
import anndata as ad
import bbknn
import gseapy as gp

#plotting
from plotnine import *
import matplotlib as mp
%matplotlib inline
```

# create AnnData obj

```python
# read raw data (expression data) to pandas df
pre_adata = pd.read_csv('/home/ubuntu/data/DL20181011_melanocyte_test_data/primary_mel_rawdata_181011.csv')

# read annotation data
anno = pd.read_csv('/home/ubuntu/data/DL20181011_melanocyte_test_data/primary_mel_metadata_181011.csv')

# pd df to np array
array_adata = pre_adata.values

# drop first column containing gene names
array_adata = array_adata[:,1:]
```

```python
# print annotation features
list(anno)
```

```python
# extract obs and var
obs = pre_adata.columns[1:]

gene_names = pre_adata.iloc[:,0].tolist()
var = pd.DataFrame()
var['gene_symbols'] = gene_names
```

```python
# create ad obj
adata = ad.AnnData(X=array_adata).T
adata.X = sparse.csr_matrix(adata.X)
adata.var_names = var.gene_symbols.values
adata.var_names_make_unique()
adata.obs_names = obs

# append metadata of choice
adata.obs['age'] = anno.age.values
adata.obs['plate'] = anno.plate.values
adata.obs['location'] = anno.Anatomical_location.values
adata.obs['race'] = anno.Race.values
adata.obs['sex'] = anno.Sex.values
adata.obs['type'] = anno.annotation.values
adata.obs['phase'] = anno.Phase.values
adata.obs['color'] = anno.sample_color.values
adata.obs['patient'] = anno.patient_id.values
# log10 expression levels of melanocyte marker PMEL
PMEL_idx = [x for x,y in enumerate(gene_names) if (y in ['PMEL'])]
PMEL_lin = pre_adata.iloc[PMEL_idx,1:].values.reshape((4356,))
PMEL_log10 = np.log10(PMEL_lin)
PMEL_log10[np.isinf(PMEL_log10)] = 0
adata.obs['PMEL'] = PMEL_log10
# log10 expression levels of melanocyte marker PMEL
KRT5_idx = [x for x,y in enumerate(gene_names) if (y in ['KRT5'])]
KRT5_lin = pre_adata.iloc[KRT5_idx,1:].values.reshape((4356,))
KRT5_log10 = np.log(KRT5_lin)
KRT5_log10[np.isinf(KRT5_log10)] = 0
adata.obs['KRT5'] = KRT5_log10
# label melanocytes vs kerotinocytes based on zscore of expresion ratio
exp_ratio = np.divide(PMEL_lin, KRT5_lin)
exp_ratio = np.log10(exp_ratio)
ratio_mean = np.mean(exp_ratio[np.isfinite(exp_ratio)])
ratio_std = np.std(exp_ratio[np.isfinite(exp_ratio)])
exp_zscore = (exp_ratio - ratio_mean) / ratio_std
ratio_labels = np.empty(exp_zscore.shape, dtype = '|U16')
ratio_labels[exp_zscore > 1] = 'MEL'
ratio_labels[exp_zscore < -1] = 'KRT'
ratio_labels[(exp_zscore <= 1) & (exp_zscore >= -1)] = 'MIX'
ratio_labels[ratio_labels == ''] = 'nan' # fill empty elements
adata.obs['ratio_label'] = ratio_labels
# # key label for gene rank analysis
# key_labels = np.empty(exp_zscore.shape, dtype = '|U16')
# key_labels[(ratio_labels == 'MEL') & ]
# adata.obs['gene_rank_key'] = 

# subset anndata obj, removinga all rows associated with ERCC
ERCC_hits = list(filter(lambda x: 'ERCC' in x, gene_names))
adata = adata[:, [x for x in gene_names if not (x in ERCC_hits)]]
```

```python
# filter ad obj based on min genes and min counts cutoff
sc.pp.filter_cells(adata, min_counts= 50000)
sc.pp.filter_cells(adata, min_genes=500)
sc.pl.highest_expr_genes(adata, n_top=20)
```

```python
# filter genes by dispersion
filter_result = sc.pp.filter_genes_dispersion(
    adata.X, min_mean=0.0125, max_mean=5, min_disp=0.5)
sc.pl.filter_genes_dispersion(filter_result)
```

```python
# PCA
adata = adata[:, filter_result.gene_subset]
sc.pp.log1p(adata)
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata)

adata.obsm['X_pca'] *= -1  # multiply by -1 to match Seurat
sc.pl.pca_variance_ratio(adata, log=True)
```

```python
# Cluster and project using select PC's
num_pcs = 20

sc.pp.neighbors(adata,n_pcs=num_pcs, n_neighbors=30)

# UAMP: Uniform Maniford Approximation and Projection
sc.tl.umap(adata)

# Louvain community clustering
sc.tl.louvain(adata, resolution = 0.9)

# plot umap
sc.pl.umap(adata, color=['louvain',
                         'age',
                         'plate',
                         'location',
                         'sex',
                         'race',
                         'type',
                         'phase',
                         'color',
                         'patient',
                         'PMEL',
                         'KRT5',
                         'ratio_label'])

#clear out the cluster plot colour map by hand, as otherwise it stays
del adata.uns['louvain_colors']
```

```python
# bbknn normalization
adata_bbknn = bbknn.bbknn(adata, 
                          batch_key='plate', 
                          neighbors_within_batch=5, 
                          n_pcs=num_pcs, 
                          save_knn=True, 
                          copy=True,
                          trim=75)


sc.tl.umap(adata_bbknn)
sc.tl.louvain(adata_bbknn, resolution = 0.9)
sc.pl.umap(adata_bbknn, color=['louvain',
                                 'age',
                                 'plate',
                                 'location',
                                 'sex',
                                 'race',
                                 'type',
                                 'phase',
                                 'color',
                                 'patient',
                                 'PMEL',
                                 'KRT5',
                                 'ratio_label'])
```

```python
warnings.filterwarnings('ignore')
for val in ['Adult','Neonatal','Fetal']:
    # Rank genes
    print('\nKey = {}'.format(val))
    idx_bool = (adata.obs['age'] == val)
    idx_bool = list(idx_bool.values)
    idx = [x for x,y in enumerate(idx_bool) if y == True]
    adata_tmp = adata[idx,:]
    sc.tl.rank_genes_groups(adata_tmp, 'ratio_label', method='logreg')
    sc.pl.rank_genes_groups(adata_tmp, n_genes=20, sharey=False)
    print(pd.DataFrame(adata_tmp.uns['rank_genes_groups']['names']).head(10))
    
    # GO term
    for cell in ['KRT', 'MEL']:
        print('\nCell = {}'.format(cell))
        glist = pd.DataFrame(adata_tmp.uns['rank_genes_groups']['names']).head(10)[cell].squeeze().tolist()

        enr = gp.enrichr(gene_list = glist,
                        description='test',
                        gene_sets=['KEGG_2016',
                                   'GO_Molecular_Function_2018',
                                   'GO_Biological_Process_2018'],
                        outdir='/home/ubuntu/data/enrichr_kegg',
                        cutoff=0.5)

        print(enr.res2d.head(10))
```
