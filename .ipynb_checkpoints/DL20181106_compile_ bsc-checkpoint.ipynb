```python
import pandas as pd
import os
```

# Read in BSC files
- Check column names
- Hard-code column names

```python
outfn = 'DL20181106_bsc_metadata.csv'
wkdir = '/home/ubuntu/data/DL20181011_melanocyte_test_data'
s3dir = 'daniel.le-work/MEL_project'

bsc_dir = '/home/ubuntu/data/DL20181011_melanocyte_test_data/bsc_data'
bsc_files = os.listdir('{}/'.format(bsc_dir))

# check colnames of uniformity
col_df = pd.DataFrame()
for file in bsc_files:
    fn = '{}/{}'.format(bsc_dir, file)
    col_df[file] = pd.read_csv(fn).columns

col_df.apply(lambda x: all(val==x[0] for val in x), axis=1)
col_df.iloc[8,:] # anomaly detected = B003063_Index.csv

# requires hard-coding column names
colnames = ['time','fsc_a','fsc_h','fsc_w','bsc_a','bsc_h','bsc_w',
            'dapi','alpha6','ckit','cd11','well','plate_barcode']
compiled_df = pd.DataFrame()
for idx,file in enumerate(bsc_files):
    fn = '{}/{}'.format(bsc_dir, file)
    plate_id = file.split('_')[0]
    bsc_df = pd.read_csv(fn)
    bsc_df['plate_barcode'] = plate_id
        
    # update colnames
    bsc_df.columns = colnames
    # compile sub dfs
    compiled_df = compiled_df.append(bsc_df)

compiled_df.to_csv('{}/{}'.format(wkdir, outfn), index=False)
! aws s3 cp --quiet {wkdir}/{outfn} s3://{s3dir}/{outfn}
    
compiled_df.head()
```
